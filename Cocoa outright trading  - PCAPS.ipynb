```python
# Imports
import numpy as np
import pandas as pd
import datetime as dt
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
from IPython import get_ipython
import matplotlib.mlab as mlab
import itertools 
import NormalDataParserPy
import sklearn
```

```python
# Inputs

marketDataCpp = NormalDataParserPy.orderBooksWithDepth()
marketDataCpp.query_filter.exchange_vec = ['ICE_CHC'];
marketDataCpp.query_filter.date_vec = ['02202018']
marketDataCpp.query_filter.symbols_vec = ['CC  FMK0018!']
marketDataPy = marketDataCpp.get(num_of_levels = 5, numOrderIdsPerLevel = 500)


t_time = pd.DatetimeIndex(pd.to_datetime(marketDataPy.protocol_message_time)).tz_localize('UTC').tz_convert('US/Central').tz_localize(None)
print 'TOTAL: ' + str(len(t_time)) + ' events'

raw_data_US = pd.DataFrame({'t_time' : t_time,
    'bid_orders' : marketDataPy.bid_orders[:,0], 'bid_size': marketDataPy.bid_size[:,0], 'bid_price': marketDataPy.bid_price[:,0],
    'ask_price': marketDataPy.ask_price[:,0], 'ask_size': marketDataPy.ask_size[:,0], 'ask_orders' : marketDataPy.ask_orders[:,0],
    'trade_size': marketDataPy.trade_size},
    columns = ['t_time', 'trade_size', 'bid_orders', 'bid_size', 'bid_price', 'ask_price', 'ask_size', 'ask_orders'])

```

```python
# data cleaning
# bad data in front
TIMESTAMP_INDEX = 1
raw_data_US.iloc[TIMESTAMP_INDEX:].tail(10)

```

```python
#data_US=raw_data_US.iloc[5500:]
#Select hours
data_US=raw_data_US.dropna(axis=0, how='any')
data_US=data_US[data_US.t_time.dt.strftime('%H:%M:%S').between('05:00:00','10:54:59')]
data_US.bid_price=data_US.bid_price/10
data_US.ask_price=data_US.ask_price/10
data_US=data_US.rename(columns={'trade_size':'trade_size_us','bid_orders':'bid_orders_us',
                                'bid_size':'bid_size_us','bid_price':'bid_price_us', 
                                'ask_orders':'ask_orders_us','ask_size':'ask_size_us',
                                'ask_price':'ask_price_us', })
```
